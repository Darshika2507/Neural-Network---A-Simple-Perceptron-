**Q.1 What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.**

      --> Deep Learning (DL) is a subset of Machine Learning (ML) using multi-layered Artificial Neural Networks (ANNs) to automatically learn complex patterns from vast, often unstructured data, unlike traditional ML, which needs manual feature engineering and works better with structured, smaller datasets. It evolved by leveraging increased data availability and powerful GPUs to build brain-inspired networks, enabling breakthroughs in tasks like image/speech recognition, but requires more data and computing power than traditional methods.

      Evolution

      Early Concepts (1940s-1960s): The idea of neural networks began with McCulloch-Pitts neurons, but early models faced limitations like the XOR problem.

      AI Winters & Resurgence (1970s-1990s): Progress stalled, but concepts like backpropagation emerged, eventually allowing neural networks to learn.

      Deep Learning Era (2000s-Present): The advent of "big data," powerful Graphics Processing Units (GPUs) for parallel processing, and refined algorithms (like Convolutional Neural Networks - CNNs) ignited the deep learning revolution, showing superior results in computer vision, NLP, and more.

      How it Differs from Traditional Machine Learning

      Feature Extraction: Traditional ML requires humans to manually select important features (e.g., edges in an image); DL automatically learns features hierarchically through its deep layers (e.g., layer 1 learns edges, layer 2 learns shapes).

      Data Needs: DL thrives on massive datasets (Big Data) for optimal performance, while traditional ML can work effectively with smaller, structured data.

      Performance with Data: Traditional ML performance plateaus with more data; DL performance generally improves with more data.

      Hardware: Traditional ML runs on standard CPUs; DL often requires high-performance GPUs/TPUs due to complex computations.

      Complexity & Interpretability: DL models (deep neural networks) are complex "black boxes," harder to interpret, whereas simpler ML models are often more transparent.

      Training: DL takes much longer to train (days/weeks) but achieves high accuracy on complex tasks; ML trains faster (seconds/hours) for simpler tasks.


**Q.2 Explain the basic architecture and functioning of a Perceptron. What are its limitations?**

      --> A Perceptron is a foundational algorithm in machine learning, a type of linear classifier. It is the simplest form of an artificial neural network, designed to perform binary classification tasks [1].

      Basic Architecture

      The architecture of a single Perceptron consists of several key components:

      Inputs: The Perceptron receives multiple input signals (features), denoted as \(x_{1},x_{2},\dots ,x_{n}\).

      Weights: Each input is associated with a specific weight, \(w_{1},w_{2},\dots ,w_{n}\), which determines the importance of that input signal.
      
      Weighted Sum (Net Input Function): The Perceptron calculates a weighted sum of its inputs and weights: \(\sum _{i=1}^{n}(x_{i}\cdot w_{i})\).
      
      Bias: A bias (\(b\)) is added to the weighted sum. The bias is an adjustable constant that allows the decision boundary to be shifted. The total input is \(z=\sum _{i=1}^{n}(x_{i}\cdot w_{i})+b\) [1].
      
      Activation Function: The total input (\(z\)) is passed through an activation function, usually a simple step function, which determines the output.

      Functioning

      The Perceptron operates in two phases: forward propagation and learning via the Perceptron learning rule.

      1. Forward Propagation (Prediction)

      The Perceptron makes a decision based on the result of the activation function [1]. The output (\(\^{y}\)) is typically binary (e.g., 0 or 1, or -1 or 1), defined by the step function:

      If \(z>\text{threshold}\), the output is 1 (or positive class).
      
      If \(z\le \text{threshold}\), the output is 0 (or negative class).

      2. Learning (Training)

      The Perceptron is trained using labeled examples. The learning rule iteratively adjusts the weights and bias to minimize errors [1].

      If the Perceptron classifies an example correctly, the weights remain unchanged.
      
      If it classifies incorrectly, the weights are updated using the Perceptron update rule:
      
      \(w_{\text{new}}=w_{\text{old}}+\alpha \cdot (\text{actual\ output}-\text{predicted\ output})\cdot x\)
      
      \(b_{\text{new}}=b_{\text{old}}+\alpha \cdot (\text{actual\ output}-\text{predicted\ output})\)

      Here, \(\alpha \) is the learning rate, a small positive constant [1]. This process guarantees convergence to a correct solution if the data is linearly separable [1].

      Limitations

      Despite its simplicity and foundational role, the Perceptron has significant limitations:

      Linear Separability: The most critical limitation is that a single-layer Perceptron can only solve problems that are linearly separable [1, 2]. It cannot learn complex, non-linear relationships. A classic example it cannot solve is the XOR (exclusive OR) logic problem [1].

      Output Type: The basic Perceptron only produces binary outputs (classifications). It cannot predict continuous numerical values (regression) or provide probability estimates [1].

      Sensitivity to Data Scaling: The performance and convergence of the Perceptron are highly dependent on the scaling of the input features [2].

      Overfitting: A basic Perceptron can be prone to overfitting, especially when dealing with high-dimensional data or a limited number of training examples [2].

      Lack of Hidden Layers: The absence of hidden layers limits its ability to learn complex hierarchical patterns and representations from data, which is a key feature of modern deep neural networks [1, 2].

**Q.4 What is the difference between Loss function and Cost function in neural networks? Provide examples.**

      --> In neural networks, a Loss Function quantifies the error for a single data point (e.g., one image), while a Cost Function aggregates these individual losses, usually as an average, across the entire dataset or a batch, providing an overall measure of model performance that is minimized during training. The cost function often includes regularization terms to prevent overfitting, making it a more general measure than the simple loss for one example, with common examples being Mean Squared Error (MSE) for regression and Cross-Entropy for classification.

      Loss Function vs. Cost Function

      Loss Function (Error Function): Measures the penalty for a single training example, indicating how far the prediction is from the true value for that specific point.

     Cost Function (Objective Function): Calculates the average loss over the entire dataset or a mini-batch, giving a comprehensive metric for model evaluation and guiding optimization algorithms like gradient descent.

     Examples

     1. Mean Squared Error (MSE) / Squared Error Loss

     Loss (for one point): \(L(y,\^{y})=(y-\^{y})^{2}\), where \(y\) is the true value and \(\^{y}\) is the prediction.
     
     Cost (average over data): \(J(\theta )=\frac{1}{m}\sum _{i=1}^{m}(y^{(i)}-\^{y}^{(i)})^{2}\) (average of squared errors for \(m\) samples).
     
     2. Binary Cross-Entropy (Log Loss)
     
     Loss (for one point): \(L(y,\^{y})=-(y\log (\^{y})+(1-y)\log (1-\^{y}))\) (measures error for binary classification).
     
     Cost (average over data): \(J(\theta )=-\frac{1}{m}\sum _{i=1}^{m}[y^{(i)}\log (\^{y}^{(i)})+(1-y^{(i)})\log (1-\^{y}^{(i)})]\) (average binary cross-entropy).

     3. Categorical Cross-Entropy

     Loss (for one point): Used for multi-class classification, measuring the difference between predicted probability distribution and true one-hot encoded label.
     
     Cost: The average of these losses across all samples, often with added regularization (like \(L1\) or \(L2\)) to form the overall cost function \(J(\theta )\).

     In essence, you calculate the loss for each sample and then average or sum them up to get the cost, which is the value you aim to minimize to make your neural network perform better.

**Q.5 What is the role of optimizers in neural networks? Compare Gradient
Descent, Adam, and RMSprop.**

      --> Optimizers in neural networks adjust model weights to minimize the loss function, with Gradient Descent being basic, while RMSprop adapts learning rates per parameter using squared gradients, and Adam combines RMSprop's adaptive rates with momentum for faster, robust convergence, making them superior for complex deep learning tasks.

      Role of Optimizers

      Minimize Loss: Find the best parameters (weights/biases) to reduce the error (loss) between predicted and actual outputs.

      Navigate Loss Landscape: Guide the model's updates down the steepest slope (gradient) of the loss function.

      Adapt Learning Rates: Adjust how much parameters change per update, improving speed and stability, especially in complex networks.

      Comparison: Gradient Descent vs. RMSprop vs. Adam

**Q.6 Write a Python program to implement a single-layer perceptron from
scratch using NumPy to solve the logical AND gate.**


```python
import numpy as np
import matplotlib.pyplot as plt

# Input data for AND gate
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Expected output
y = np.array([0, 0, 0, 1])

# Initialize weights and bias
weights = np.random.rand(2)
bias = np.random.rand()

# Learning rate
lr = 0.1

# Step activation function
def step_function(x):
    return 1 if x >= 0 else 0

# Training
epochs = 20
errors = []

for epoch in range(epochs):
    total_error = 0

    for i in range(len(X)):
        # Weighted sum
        z = np.dot(X[i], weights) + bias

        # Prediction
        y_pred = step_function(z)

        # Error
        error = y[i] - y_pred
        total_error += abs(error)

        # Weight and bias update
        weights += lr * error * X[i]
        bias += lr * error

    errors.append(total_error)
    print(f"Epoch {epoch+1}, Total Error: {total_error}")

# Testing the perceptron
print("\nTesting AND Gate:")
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    output = step_function(z)
    print(f"Input: {X[i]} Output: {output}")

# Plot training error
plt.plot(errors, marker='o')
plt.xlabel("Epochs")
plt.ylabel("Total Error")
plt.title("Perceptron Training Error for AND Gate")
plt.grid()
plt.show()
```

**Q.7 Implement and visualize Sigmoid, ReLU, and Tanh activation functions
using Matplotlib.**


```python
import numpy as np
import matplotlib.pyplot as plt

# Input range
x = np.linspace(-10, 10, 400)

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plot the activation functions
plt.figure(figsize=(8, 6))

plt.plot(x, y_sigmoid, label="Sigmoid", linewidth=2)
plt.plot(x, y_relu, label="ReLU", linewidth=2)
plt.plot(x, y_tanh, label="Tanh", linewidth=2)

plt.xlabel("Input (x)")
plt.ylabel("Output")
plt.title("Activation Functions Visualization")
plt.legend()
plt.grid(True)
plt.show()
```

**Q.8 Use Keras to build and train a simple multilayer neural network on the
MNIST digits dataset. Print the training accuracy.**


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize pixel values
X_train = X_train / 255.0
X_test = X_test / 255.0

# Convert labels to one-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build the neural network model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=32,
    verbose=1
)

# Print training accuracy
training_accuracy = history.history['accuracy'][-1]
print("\nTraining Accuracy:", training_accuracy)
```

**Q.9 Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.**

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical

# Load Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalize the data
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build the neural network
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=10,
    batch_size=32,
    verbose=1
)

# Plot Accuracy
plt.figure()
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend(['Training Accuracy', 'Validation Accuracy'])
plt.grid()
plt.show()

# Plot Loss
plt.figure()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend(['Training Loss', 'Validation Loss'])
plt.grid()
plt.show()
```

**Q.10 You are working on a project for a bank that wants to automatically detect fraudulent transactions. The dataset is large, imbalanced, and contains structured features like transaction amount, merchant ID, and customer location. The goal is to classify each transaction as fraudulent or legitimate.**

**Explain your real-time data science workflow:**

** How would you design a deep learning model (perceptron or multilayer NN)?**

** Which activation function and loss function would you use, and why?**

** How would you train and evaluate the model, considering class imbalance?**

** Which optimizer would be suitable, and how would you prevent overfitting?**

     --> Real-Time Fraud Detection  Data Science Workflow

     Fraud detection in banking is a binary classification problem where each transaction must be classified as fraudulent (1) or legitimate (0).

     Since the dataset is large, imbalanced, and arrives in real time, the workflow must be accurate, fast, and robust.

     1 Designing the Deep Learning Model

     Choice of Model: Multilayer Neural Network (MLP)

     A single-layer perceptron is too simple and cannot capture complex fraud patterns.

     Fraud depends on non-linear relationships between:

     Transaction amount

     Merchant ID

     Customer location

     Transaction frequency and time patterns

     Model Architecture (Example)

     Input Layer

     Takes structured numerical features (after normalization and encoding).

     Hidden Layers (23 layers)

     Dense layers with decreasing neurons (e.g., 128  64  32).

     Output Layer

     One neuron for binary classification (fraud / non-fraud).

      This architecture balances accuracy and real-time speed.

     2 Activation Function and Loss Function

      Activation Functions

     Hidden Layers  ReLU

     Handles non-linear patterns effectively.

     Computationally fast (important for real-time systems).

     Avoids vanishing gradient problem.

     Output Layer  Sigmoid

     Outputs probability between 0 and 1.

     Ideal for binary classification.

     Loss Function: Binary Cross-Entropy

     Loss = [log()+(1)log(1)]

     Why?

     Penalizes wrong fraud predictions heavily.

     Works well with probabilistic outputs.

     Standard choice for fraud detection problems.

     3 Training and Evaluation (Handling Class Imbalance)

     Challenge: Class Imbalance

     Fraud transactions are very rare (often <1%).

     Accuracy alone is misleading.

      Techniques to Handle Imbalance

     Class Weighting

     Assign higher weight to fraud class during training.

     Prevents model from ignoring minority class.

     Resampling

     Oversampling fraud (SMOTE)

     Undersampling legitimate transactions

     Threshold Tuning

     Use probability threshold lower than 0.5 to catch more fraud cases.

      Evaluation Metrics (More Important than Accuracy)

     Precision  How many flagged frauds are actually fraud

     Recall (Very Important)  How many frauds are correctly detected

     F1-Score  Balance between precision and recall

     ROC-AUC  Overall model discrimination ability

      Recall is prioritized because missing fraud is costly.

     4 Optimizer and Overfitting Prevention

      Optimizer: Adam

     Why Adam?

     Adaptive learning rate.

     Faster convergence on large datasets.

     Works well for noisy and sparse data.

      Preventing Overfitting

     Dropout Layers

     Randomly deactivate neurons.

     Prevents memorization of data.

     Early Stopping

     Stop training when validation loss stops improving.

     L2 Regularization

     Penalizes large weights.

     Encourages simpler models.

     Batch Normalization

     Stabilizes training.

     Improves generalization.

     5 Real-Time Deployment Workflow

     Incoming transaction  feature engineering

     Model predicts fraud probability

     If probability > threshold  flagged for review

     Predictions logged for continuous retraining

     Model updated periodically to handle new fraud patterns

      Final Conclusion

     A multilayer neural network with ReLU and Sigmoid activations, trained using Binary Cross-Entropy loss and Adam optimizer, is suitable for real-time fraud detection.

     Handling class imbalance using class weighting and evaluating the model using recall and ROC-AUC ensures reliable fraud detection.

     Regularization techniques and early stopping help prevent overfitting and improve real-world performance.
